```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función original f(x) = e^x
def f(x):
    return np.exp(x)

# Definir la serie de Fourier truncada
def fourier_series(x, N):
    L = 2  # Intervalo [-2,2], entonces L = 2
    a0 = (1/L) * (np.exp(L) - np.exp(-L)) / 2  # Coeficiente a0

    # Inicializar la serie de Fourier con a0/2
    f_N = a0 / 2 * np.ones_like(x)

    # Calcular coeficientes an y bn
    for n in range(1, N+1):
        an = (1/L) * ((np.exp(L) * np.cos(n*np.pi)) - (np.exp(-L) * np.cos(-n*np.pi))) / (1 + (n*np.pi/(2*L))**2)
        bn = (1/L) * ((np.exp(L) * np.sin(n*np.pi)) - (np.exp(-L) * np.sin(-n*np.pi))) / (1 + (n*np.pi/(2*L))**2)

        # Sumar los términos a la serie
        f_N += an * np.cos(n * np.pi * x / L) + bn * np.sin(n * np.pi * x / L)

    return f_N

# Rango de valores para x
x = np.linspace(-2, 2, 400)

# Número de términos en la serie de Fourier
N = 10  # Puedes aumentar este valor para mayor precisión

# Calcular la serie de Fourier con N términos
f_approx = fourier_series(x, N)

# Graficar la función original y su aproximación de Fourier
plt.plot(x, f(x), label=r'$f(x) = e^x$', color='black', linestyle='dashed')
plt.plot(x, f_approx, label=f'Serie de Fourier con N={N}', color='blue')

# Configuración del gráfico
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.title(f'Aproximación de $e^x$ con Serie de Fourier (N={N})')
plt.legend()
plt.grid()
plt.show()
```
