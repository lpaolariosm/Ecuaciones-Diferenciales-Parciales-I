<font color=pink>**Gràfica de una parabola**</font>

Esta es una gràfica de una parabola que tiene por ecuaciòn $y = x^2$, y es una prueba para aprender como funciona esta plataforma.



```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)  # Valores de x
y = x**2  # Función y = x^2

plt.plot(x, y)  # Graficar
plt.xlabel('x')
plt.ylabel('y')
plt.title('$y = x^2$')
plt.grid()
plt.show()
```
