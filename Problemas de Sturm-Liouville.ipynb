## Resolver la ecuación diferencial

$X''+ λX=0 \quad X(0)=X(1)=0$

```python
from sympy import *
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

Definir las variables

```python
x = sp.symbols('x')
X = sp.Function('X')
k = sp.pi**2  # Valor de k
C1, C2 = sp.symbols('C1 C2')
```

$X'' + kX = 0$

```python
ode1= sp.Eq(X(x).diff(x, 2) + k*X(x), 0)
```

Solucion general

```python
sol1 = sp.dsolve(ode1, X(x))
Xsol1 = sol1.rhs
sol1
```

Aplicar condiciones: X(0) = 0, X(1) = 0

```python
conds1 = [
    Xsol1.subs(x, 0),
    Xsol1.subs(x, 1)
]
sol_c1 = sp.solve(conds1, (C1, C2))
Xsol1 = Xsol1 = Xsol1.subs(sol_c1).subs(C1, 1) # C2 es la constante libre

Xsol1
```

```python
func1 = sp.lambdify(x, Xsol1, 'numpy')
xx = np.linspace(0, 1, 200)
yy1 = func1(xx)
```

```python
plt.plot(xx, yy1, 'r', label='Problema 1', color='red')
plt.ylabel('X(x)')
plt.title('Solución del Problema 1: X(0)=0, X(1)=0')
plt.grid(True)
plt.legend()
plt.show()
```

## Resolver la ecuación diferencial

$X''+ λX=0 \quad X'(0)=X'(1)=0$

```python
from sympy import *
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = sp.symbols('x')
X = sp.Function('X')
k = sp.pi**2  # Valor ejemplo de k
C1, C2 = sp.symbols('C1 C2')
```

```python
ode2 =  sp.Eq(X(x).diff(x, 2) + k*X(x), 0)
```

Solución general

```python
sol2 =  sp.dsolve(ode2, X(x))
Xsol2 = sol2.rhs
dXsol2 = Xsol2.diff(x)
sol2
```

 Aplicar condiciones: X'(0) = 0, X'(1) = 0

```python
const2= [
    dXsol2.subs(x, 0),  # X'(0) = 0
    dXsol2.subs(x, 1)   # X'(1) = 0
]
sol_c2 = sp.solve(const2, (C1, C2)) # Changed 'conds2' to 'const2'
# Sustituir soluciones y asignar valor a la constante restante
Xsol2 = Xsol2.subs(sol_c2).subs(C2, 1)  # C1 es la constante libre
Xsol2
```

```python
func2 = sp.lambdify(x, Xsol2, 'numpy')
xx = np.linspace(0, 2, 200)
yy2 = func2(xx)
```

```python
plt.plot(xx, yy2, 'b', label='Problema 2', color='orange')
plt.xlabel('x')
plt.ylabel('X(x)')
plt.title("Solución del Problema 2: X'(0)=0, X'(1)=0")
plt.grid(True)
plt.legend()
plt.show()
```
