# <span style="color:red;"> " ONDAS VIAJERAS "</span>
<span style="color:orange;">**$f(x)= \exp (-\frac{(x'-0.5)^2}{0.01})$**</span>

<span style="color:grey"> (a) Mostrar que al hacer el cambio de variable $x'=x-ct$ la perturbación se desplaza a la derecha </span>

<span style="color:grey"> (b) Mostrar que al hacer el cambio de variable $x'=x+ct$ la perturbación se desplaza a la izquierda </span>

<span style="color:grey"> (c) Graficar la superposición de las ondas viajeras de los incisos (a) y (b)</span>

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definimos las variables simbólicas
x, t, c = sp.symbols('x t c')

# Función perturbación original: f(x') = exp(-(x'-0.5)^2 / 0.01)
x_prime = sp.symbols("x'")
f = sp.exp(-((x_prime - 0.5)**2) / 0.01)
```

```python
# (a) Cambio de variable x' = x - c*t (onda viajera a la derecha)
x1 = x - c*t
f_a = f.subs(x_prime, x1)
print("Onda viajera a la derecha (a):")
sp.pprint(f_a)
```

```python
# (b) Cambio de variable x' = x + c*t (onda viajera a la izquierda)
x2 = x + c*t
f_b = f.subs(x_prime, x2)
print("\nOnda viajera a la izquierda (b):")
sp.pprint(f_b)
```

```python
# (c) Superposición de las ondas viajeras
f_c = f_a + f_b
print("\nSuperposición de ondas viajeras (c):")
sp.pprint(f_c)
```

```python
# Graficar las tres funciones para un tiempo t fijo y c=1
f_a_func = sp.lambdify((x, t), f_a.subs(c, 1), 'numpy')
f_b_func = sp.lambdify((x, t), f_b.subs(c, 1), 'numpy')
f_c_func = sp.lambdify((x, t), f_c.subs(c, 1), 'numpy')
```

```python
# Definir valores de x y t para graficar
x_vals = np.linspace(-1, 2, 500)
t_val = 0.2  # Tiempo fijo
```

```python
# Evaluar las funciones
y_a = f_a_func(x_vals, t_val)
y_b = f_b_func(x_vals, t_val)
y_c = f_c_func(x_vals, t_val)
```

```python
# Graficar
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_a, label='(a) Onda a la derecha', color='orange')
plt.plot(x_vals, y_b, label='(b) Onda a la izquierda', color='yellow')
plt.plot(x_vals, y_c, label='(c) Superposición', color='red', linestyle='--')
plt.title("Ondas viajeras y su superposición")
plt.xlabel("x")
plt.ylabel("f(x,t)")
plt.legend()
plt.grid(True)
plt.show()
```
