# <span style="color:red;"> " ONDAS VIAJERAS "</span>
<span style="color:orange;">**$f(x)= \exp (-\frac{(x'-0.5)^2}{0.01})$**</span>

<span style="color:grey"> (a) Mostrar que al hacer el cambio de variable $x'=x-ct$ la perturbación se desplaza a la derecha </span>

<span style="color:grey"> (b) Mostrar que al hacer el cambio de variable $x'=x+ct$ la perturbación se desplaza a la izquierda </span>

<span style="color:grey"> (c) Graficar la superposición de las ondas viajeras de los incisos (a) y (b)</span>

```python
import numpy as np
import matplotlib.pyplot as plt
```

Parametros

```python
c = 1.0     # Velocidad de la onda
t = 0.2     # Tiempo en segundos
x = np.linspace(0, 1, 1000)  # Dominio espacial
```

Definición de la función gaussiana original en x'

```python
def f(xp):
    return np.exp(-((xp - 0.5)**2) / 0.01)

# Inciso (a): onda viajando a la derecha: x' = x - ct
f_derecha = f(x - c * t)

# Inciso (b): onda viajando a la izquierda: x' = x + ct
f_izquierda = f(x + c * t)

# Inciso (c): superposición de ambas ondas
f_superpuesta = f_derecha + f_izquierda
```

Gráfica para inciso (a)

```python
plt.figure(figsize=(8, 4))
plt.plot(x, f_derecha, color='blue')
plt.title("Inciso (a): Onda viajando a la derecha (x' = x - ct)")
plt.xlabel("x")
plt.ylabel("f(x, t)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

 Gráfica para inciso (b)

```python
plt.figure(figsize=(8, 4))
plt.plot(x, f_izquierda, color='green')
plt.title("Inciso (b): Onda viajando a la izquierda (x' = x + ct)")
plt.xlabel("x")
plt.ylabel("f(x, t)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

Gráfica para inciso (c)

```python
plt.figure(figsize=(8, 4))
plt.plot(x, f_superpuesta, color='red')
plt.title("Inciso (c): Superposición de ambas ondas")
plt.xlabel("x")
plt.ylabel("f(x, t)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

Las 3 ondas en un solo grafico

```python
# Inciso (a): onda viajando a la derecha: x' = x - ct
f_derecha = f(x - c * t)

# Inciso (b): onda viajando a la izquierda: x' = x + ct
f_izquierda = f(x + c * t)

# Inciso (c): superposición de ambas ondas
f_superpuesta = f_derecha + f_izquierda

# Gráficas
plt.figure(figsize=(10, 6))

# Onda viajando a la derecha
plt.plot(x, f_derecha, label="Onda viajando a la derecha (x' = x - ct)", color='blue')

# Onda viajando a la izquierda
plt.plot(x, f_izquierda, label="Onda viajando a la izquierda (x' = x + ct)", color='green')

# Superposición
plt.plot(x, f_superpuesta, label="Superposición", color='red', linestyle='--')

plt.title("Ondas viajeras y su superposición")
plt.xlabel("x")
plt.ylabel("f(x, t)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
