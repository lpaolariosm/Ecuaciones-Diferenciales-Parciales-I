```python
from sympy import fourier_series, pi, plot, Piecewise
from sympy.abc import x
```

```python
f=Piecewise(
    (0, x < -2),
    (x+2,(x >= -2) & (x < 0)),
    (0,(x<=2)))
```

```python
plot(f, (x,-2,2),line_color ='pink')
```

```python
s = fourier_series(f, (x, -2, 2))
s1 = s.truncate(n = 3)
s2 = s.truncate(n = 5)
s3 = s.truncate(n = 7)
p = plot(f, s1, s2, s3, (x, -pi, pi), show=False, legend=True)
```

```python
s
```

```python
p[0].label = 'x'
p[1].label = 'n=3'
p[2].label = 'n=5'
p[3].label = 'n=7'
```

```python
p.show()
```

```python
f=Piecewise((exp(2*x),(x>=0) & (x<=1)),(-exp(-2*x),(x>=-1) & (x<=0)))
```

Coeficientes de la serie de Fourier en senos

```python
s = fourier_series(f, (x, -1, 1))
s1 = s.truncate(n = 5)
s2 = s.truncate(n = 7)
s3 = s.truncate(n = 9)
p = plot(f, s1, s2, s3, (x, -pi, pi), show=False, legend=True)
```

```python
plot(f, (x,0,pi),line_color ='purple')
```
